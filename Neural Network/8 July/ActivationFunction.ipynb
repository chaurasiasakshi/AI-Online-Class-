# Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,7))
```

```python
def sigmoid_b(x):
    return(1/(1+np.exp(-x)))

# binary function use in - output layer
```

```python
x = np.linspace(-5,5)
plt.plot(x , sigmoid_b(x))
plt.title('Activation Binary Sigmoid Function')
```

```python
def TanH(x):
    return np.tanh(x)


plt.plot(x , TanH(x))
plt.title('Activation TanH Function')
plt.show()
```

```python
def RectifiedLinearUnit(x):
    list=[]
    for i in x :
        if i>=0:
            list.append(i)
        else:
            list.append(0)
    return list

x=np.linspace(-15,5)
plt.plot(x,RectifiedLinearUnit(x))
plt.title('RectifiedLinearUnit Function')
plt.show()
```

```python
def Leakyrelu(x):
    list =[]
    for i in x:
        if i>=0:
            list.append(i)
        else:
            list.append(0.01*i)
    return list

plt.plot(x,Leakyrelu(x))
plt.title('LeakyRelu Function')
plt.show()
```

```python
x=np.linspace(-15,5)
plt.plot(x,Leakyrelu(x))
plt.title('LeakyRelu Function')
plt.show()
```

```python
def softmax(x):
    return (np.exp(x)/np.sum(np.exp(x)))

#x=np.linspace(-15,5)
plt.plot(x,softmax(x))
plt.title('SoftMax Function')
plt.show()
```
